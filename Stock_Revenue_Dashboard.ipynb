# Question 5: Tesla Stock Price Graph

```python
# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Create Ticker object for Tesla and extract data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # Use 1 year as a fallback if "max" fails
tesla_data.reset_index(inplace=True)

# Define make_graph function
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Tesla stock data
make_graph(tesla_data, title="Tesla Stock Price Over Time")
```
